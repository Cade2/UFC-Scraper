```python
from curl_cffi import requests
import os
from pydantic import BaseModel
from rich import print
```

```python
class SearchEvents(BaseModel):
    eventID: str
    eventName: str
    eventDate: str
    eventURL: str
    eventStatus: str 
    eventCountry: str
    eventCity: str
    venueName: str
    eventTime: str
```

```python
class SearchResponse(BaseModel):
    count: int
    searchTerm: str
    items: list[SearchEvent]
```

```python
class EventDetails(BaseModel):
    eventID: int
    eventName: str
    fightID: int
    fightOrder: int
    fightStatus: str
    cardSegement: str
    fighterName: str
    birthPlace: str
    fightingOutOf: str
    Record: str
    DOB: str
    age: int
    stance: str
    weight: float
    height: float
    reach: float
    fighterURL: str
    weightClasses: str
    titleFight: str
    corner: str
    rounds: int
    roundFormat: str
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import time

# Set up Chrome WebDriver with DevTools open
options = webdriver.ChromeOptions()
options.add_argument("--auto-open-devtools-for-tabs")  # Opens DevTools automatically
options.add_experimental_option("detach", True)  # Keeps the browser open
driver = webdriver.Chrome(options=options)

# Step 1: Open UFC Events Page
driver.get("https://www.ufc.com/events#events-list-upcoming")
time.sleep(5)  # Wait for the page to load

# Step 2: Handle Cookie Popup
try:
    cookie_button = driver.find_element(By.ID, "onetrust-accept-btn-handler")
    cookie_button.click()
    time.sleep(2)  # Wait for it to disappear
    print(" Cookie popup handled.")
except:
    print(" No cookie popup found, continuing...")

# Step 3: Open DevTools (Inspect Element)
action = ActionChains(driver)
action.key_down(Keys.CONTROL).key_down(Keys.SHIFT).send_keys('i').key_up(Keys.CONTROL).key_up(Keys.SHIFT).perform()
time.sleep(3)  # Wait for DevTools to open
print(" DevTools opened.")

# Step 4: Click the First Upcoming Event
try:
    event_link = driver.find_element(By.XPATH, "//div[contains(@class, 'c-card-event--result__info')]//a")
    event_url = event_link.get_attribute("href")  # Get event URL
    driver.get(event_url)  # Navigate to event page
    print(f" Opened Event Page: {event_url}")
    time.sleep(5)  # Wait for event page to load
except Exception as e:
    print(" Error clicking event:", e)

# Step 5: Click the ">>" button in DevTools and select Network
try:
    action = ActionChains(driver)

    # Press "Ctrl + Shift + P" to open the DevTools Command Menu
    action.key_down(Keys.CONTROL).key_down(Keys.SHIFT).send_keys('p').key_up(Keys.CONTROL).key_up(Keys.SHIFT).perform()
    time.sleep(2)  # Wait for command menu to open

    # Type "network" and press Enter
    action.send_keys("network").send_keys(Keys.ENTER).perform()
    print(" Switched to Network tab.")
except Exception as e:
    print(" Failed to switch to Network tab:", e)

print(" Go to DevTools  Network  Fetch/XHR to inspect API responses.")

# Keeps the browser open
```
